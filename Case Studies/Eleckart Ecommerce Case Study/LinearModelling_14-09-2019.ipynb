```python
print(Eleckart.product_analytic_sub_category.isnull().sum())

Eleckart['product_analytic_sub_category'].replace(r'\s+', np.nan,regex=True,inplace=True)
Eleckart=Eleckart[Eleckart['product_analytic_sub_category']!=0]

df1=Eleckart.product_analytic_sub_category.str.contains('CameraAccessory')
CameraAccessory=Eleckart[df1]

df2=Eleckart.product_analytic_sub_category.str.contains('GamingAccessory')
GamingAccessory=Eleckart[df2]

df3=Eleckart.product_analytic_sub_category.str.contains('HomeAudio')
HomeAudio=Eleckart[df3]
```

```python
#Inspecting CameraAccessory sub category data
CameraAccessory.head()
```

```python
CameraAccessory.drop(['Unnamed: 0','index'],axis=1,inplace=True)
CameraAccessory.head()
```

```python
#Inspecting columns types 
CameraAccessory.info()
```

```python
CameraAccessory_data = CameraAccessory[['Month','gmv','deliverybdays','deliverycdays','units','s1_fact.order_payment_type','sla','product_analytic_super_category','product_analytic_category','product_analytic_sub_category','product_analytic_vertical','product_mrp','product_procurement_sla','frequency_y','percent','Year_y','Total Investment','TV','Digital','Sponsorship','Content Marketing','Online marketing','Affiliates','SEM','Radio','Other','NPS','Day','Stock Index','Week','Year','sales_start_week','sales_end_week','Weekday']]
```

```python
CameraAccessory_data.info()
```

```python
#Creating dummies for all categorical variables
s1_fact_order_payment_type = pd.get_dummies(CameraAccessory_data['s1_fact.order_payment_type'])
CameraAccessory_data = pd.concat([CameraAccessory_data,s1_fact_order_payment_type], axis = 1)

product_analytic_super_category  = pd.get_dummies(CameraAccessory_data['product_analytic_super_category'], drop_first = True)
CameraAccessory_data = pd.concat([CameraAccessory_data,product_analytic_super_category], axis = 1)

product_analytic_sub_category = pd.get_dummies(CameraAccessory_data['product_analytic_sub_category'], drop_first = True)
CameraAccessory_data = pd.concat([CameraAccessory_data,product_analytic_sub_category], axis = 1)

product_analytic_category = pd.get_dummies(CameraAccessory_data['product_analytic_category'], drop_first = True)
CameraAccessory_data = pd.concat([CameraAccessory_data,product_analytic_category], axis = 1)
                            
product_analytic_vertical = pd.get_dummies(CameraAccessory_data['product_analytic_vertical'], drop_first = True)
CameraAccessory_data = pd.concat([CameraAccessory_data,product_analytic_vertical], axis = 1)

Weekday = pd.get_dummies(CameraAccessory_data['Weekday'], drop_first = True)
CameraAccessory_data = pd.concat([CameraAccessory_data,Weekday], axis = 1)

#special_sales = pd.get_dummies(CameraAccessory_data['special_sales'], drop_first = True)
#CameraAccessory_data = pd.concat([CameraAccessory_data,special_sales], axis = 1)
```

```python
#Deleting original columns
CameraAccessory_data = CameraAccessory_data.drop(['product_analytic_vertical','s1_fact.order_payment_type','product_analytic_super_category','product_analytic_category','product_analytic_sub_category'],axis=1)
#Deleting original columns
#CameraAccessory_data = CameraAccessory_data.drop(['special_sales'],axis=1)
CameraAccessory_data = CameraAccessory_data.drop(['Weekday'],axis=1)
```

```python
CameraAccessory_data.head()
```

```python
var=CameraAccessory_data.columns

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
CameraAccessory_data[var]=scaler.fit_transform(CameraAccessory_data[var])
CameraAccessory_data.head()
```

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
df_train, df_test = train_test_split(CameraAccessory_data, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
y_train = df_train.pop('gmv')
X_train = df_train
```

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 10)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()   # Running the linear model
```

```python
#Let's see the summary of our linear model
print(lm.summary())
```
