```python
#Converting categorical variables to numeric variable
varlist = ['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea']

def apply_map(x):
    return x.map({'yes':1,'no':0})

housing[varlist] = housing[varlist].apply(apply_map)
```

```python
housing.head()
```

```python
status = pd.get_dummies(housing['furnishingstatus'],drop_first=True)
```

```python
housing = pd.concat([housing,status],axis=1)
```

```python
housing.drop(['furnishingstatus'],axis=1,inplace=True)
```

```python
#Test Train split
from sklearn.model_selection import train_test_split

df_train,df_test = train_test_split(housing,train_size= 0.7,test_size = 0.3,random_state=100)
```

```python
#Scaling numerical features
var_list = ['area','bedrooms','bathrooms','stories','parking','price']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
df_train[var_list] = scaler.fit_transform(df_train[var_list])
```

```python
#plot correlation matrix as a heatmap
plt.figure(figsize = (16,12))
sns.heatmap(df_train.corr(),annot=True,cmap="YlGnBu")
plt.show()
```

```python
#plot area against price
plt.figure(figsize=(8,8))
plt.scatter(df_train.price,df_train.area)
plt.show()
```

```python
#Diving into predictor and output variables
y_train = df_train.pop('price')
```

```python
x_train = df_train
```

```python
#Building a model
import statsmodels.api as sm
x_train_lm = sm.add_constant(x_train[['area']])

lr = sm.OLS(y_train,x_train_lm).fit()
```

```python
lr.params
```

```python
#plotting predicted value against X
plt.scatter(x_train_lm.iloc[:,1],y_train)
plt.plot(x_train_lm.iloc[:,1],0.126 + 0.467 * x_train_lm.iloc[:,1],'r')
plt.show()
```

```python
print(lr.summary())
```

```python
#Adding another variable to improce the model
x_train_lm = sm.add_constant(x_train[['area','bathrooms']])
lr = sm.OLS(y_train,x_train_lm).fit()
print(lr.summary())
```

```python
#Adding another variable to improce the model
x_train_lm = sm.add_constant(x_train[['area','bathrooms','bedrooms']])
lr = sm.OLS(y_train,x_train_lm).fit()
print(lr.summary())
```

```python
#Adding all variable to improce the model
x_train_lm = sm.add_constant(x_train)
lr = sm.OLS(y_train,x_train_lm).fit()
print(lr.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
#Determining vif for all predictor variables
vif = pd.DataFrame()
vif['Features'] = x_train.columns
vif['VIF'] = [variance_inflation_factor(x_train.values,i) for i in range(x_train.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
